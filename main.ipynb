```python
# Null imputation
a = df["writers"].mode()[0]
a
```

```python
df['writers'] = df["writers"].fillna("Unknown")
```

```python
df.isna().sum()
```

```python
df['features'] = df['genres'] + " " + df['directors'] + " " + df['writers']
```

```python
df
```

```python
# encoding using tfidf vectorizer

from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(stop_words='english')
feature_matrix = vectorizer.fit_transform(df['features'])

feature_matrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(feature_matrix)

similarity
```

```python
def recommend(movie_name, top_n=5):
    
    # index of the movie
    index = df[df['primaryTitle'] == movie_name].index[0]
    
    # similarity scores of that movie
    scores = similarity[index]
    
    score_list = []
    
    for i in range(len(scores)):
        score_list.append((scores[i], i))
    
    score_list.sort(reverse=True)
    
    # top similar movies 
    for i in range(1, top_n + 1):
        movie_index = score_list[i][1]
        print(df.iloc[movie_index]['primaryTitle'])
```

```python
recommend("Inception")
```

```python
recommend("Titanic")
```

```python
recommend("The Fast and the Furious")
```
